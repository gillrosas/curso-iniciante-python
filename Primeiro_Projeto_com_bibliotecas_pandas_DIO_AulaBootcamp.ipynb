```python
#criando nosso Dataframe
df = pd.read_excel("AdventureWorks.xlsx")
```

```python
#Visualizando as 5 primeiras linhas 
df.head(5)
```

```python
#quantidade de linhas e colunas 
df.shape
```

```python
#verificando os tipos de dados
df.dtypes
```

```python
#qual a receita total 
df["Valor Venda"].sum()
```

```python
#qual o custo total?
#cria a coluna de custo total 
df["Custo Total"] = df["Custo Unitário"].mul(df["Quantidade"])
df.head(1)
```

```python
#Custo total 
round(df["Custo Total"].sum(), 2)
```

```python
#Temos a receita e custo e o total, podemos achar o lucro total 
#Vamos criar uma coluna de o Lucro que será Receita - Custo 
df["Lucro Total"] = df["Valor Venda"] - df["Custo Total"]
df.head(1)
```

```python
#total lucro 
round(df["Lucro Total"].sum(), 2)
```

```python
#tempo para enviar um produto 
df["Tempo_de_envio"] = df["Data Envio"] - df["Data Venda"]
df.head(1)
```

```python
#Extraindo apenas os dias
df["Tempo_de_envio"] = (df["Data Envio"] - df["Data Venda"]).dt.days
df.head(5)
```

```python
#Médi do tempo de envio por marcar
df.groupby("Marca")["Tempo_de_envio"].mean()
```

```python
#Verificando se possível valores ausentes
df.isnull().sum()
```

```python
#Vamos agrupar por ano e marca 
df.groupby([df["Data Venda"].dt.year, "Marca"])["Lucro Total"].sum()
```

```python
#configurar a notação cientifica
pd.options.display.float_format = "{:20,.2f}".format
```

```python
#Resetando o idex/salvar num novo excel
lucro_ano = df.groupby([df["Data Venda"].dt.year, "Marca"])["Lucro Total"].sum().reset_index()
lucro_ano
```

```python
#Qual o total de produtos vendidos?
df.groupby("Produto")["Quantidade"].sum().sort_values(ascending=False)
```

```python
#Gráfico total de produtos
df.groupby("Produto")["Quantidade"].sum().sort_values(ascending= False).plot.barh(title = "Tota de produtos vendidos", color = "Green")
plt.xlabel("Quantidade");
plt.ylabel("Produto");
```

```python
df.groupby(df["Data Venda"].dt.year)["Lucro Total"].sum().plot.bar(color = "yellow")
plt.xlabel("Ano")
plt.ylabel("Lucro")

```

```python
#selecionando apenas as vendas de 2009
df_2009= df[df["Data Venda"].dt.year == 2009]
df.head(5)
```

```python
#gráfico de linhas por mês e luco 
df_2009.groupby(df["Data Venda"].dt.month)["Lucro Total"].sum().plot(title = "Lucro X Mês")
```

```python
#gráfico da marca que está dando mais lucro 
df_2009.groupby(df["Marca"])["Lucro Total"].sum().plot.bar(title= "Lucro X Marca")
```

```python
df_2009.groupby(df["Classe"])["Lucro Total"].sum().plot.bar(title= "Lucro X Marca")
```
